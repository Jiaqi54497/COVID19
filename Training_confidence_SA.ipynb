```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_train=pd.read_csv('South_Africa_save/x_train.csv')
x_train1=pd.read_csv('China_save/x_train.csv')
```

```python
x_train=x_train.drop(['Unnamed: 0'],axis=1)
x_train1=x_train1.drop(['Unnamed: 0'],axis=1)
```

```python
for i in range(x_train.shape[1]):
    plt.scatter([i]*len(x_train[str(i)]),x_train[str(i)])
plt.ylabel('policy strength')
plt.xticks(range(x_train.shape[1]))
plt.xlabel('policy number')
plt.title('Training inputs for South Africa')
```

```python
for i in range(x_train1.shape[1]):
    plt.scatter([i]*len(x_train1[str(i)]),x_train1[str(i)])
plt.ylabel('policy strength')
plt.xticks(range(x_train1.shape[1]))
plt.xlabel('policy number')
plt.title('Training inputs for China')
```
